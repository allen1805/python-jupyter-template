### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 10)
y = x ** 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot (Top-left)
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')
axs[0, 0].set_title("Line Plot")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values")
axs[0, 0].grid(True)
axs[0, 0].legend()

# Bar chart (Top-right)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Scatter plot (Bottom-left)
axs[1, 0].scatter(scatter_x, scatter_y, color='purple', marker='o')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")
```
