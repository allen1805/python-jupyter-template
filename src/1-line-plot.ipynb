### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 11)  # x values from 0 to 10
y = x ** 2  # y values as the square of x

# Plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')

# Customizations
plt.title("Line Plot of y = x²")
plt.xlabel("X values")
plt.ylabel("Y values (X squared)")
plt.grid(True)  # Enable grid
plt.legend()  # Show legend

# Show plot
plt.show()
```
